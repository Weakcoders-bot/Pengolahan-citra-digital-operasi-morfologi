```python
# Operasi Morfologi pada Citra Biner
### Pengolahan Citra Digital  
##Universitas Pamulang

import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
os.listdir(r"C:\Penyimpanan Utama\Documents\Jupyter Notebook")
```

```python
img_path = r"C:\Penyimpanan Utama\Documents\Jupyter Notebook\unpam-logo.png"

img = cv2.imread(img_path)

if img is None:
    raise FileNotFoundError("Gambar tidak ditemukan. Periksa path atau nama file.")
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Citra Asli (Logo UNPAM)")
plt.axis("off")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
plt.title("Citra Grayscale")
plt.axis("off")
```

```python
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(binary, cmap="gray")
plt.title("Citra Biner")
plt.axis("off")
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
kernel
```

```python
erosion = cv2.erode(binary, kernel, iterations=1)

plt.imshow(erosion, cmap="gray")
plt.title("Erosion")
plt.axis("off")
```

```python
dilation = cv2.dilate(binary, kernel, iterations=1)

plt.imshow(dilation, cmap="gray")
plt.title("Dilation")
plt.axis("off")
```

```python
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)

plt.imshow(opening, cmap="gray")
plt.title("Opening")
plt.axis("off")
```

```python
closing = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)

plt.imshow(closing, cmap="gray")
plt.title("Closing")
plt.axis("off")
```

```python
gradient = cv2.morphologyEx(binary, cv2.MORPH_GRADIENT, kernel)

plt.imshow(gradient, cmap="gray")
plt.title("Morphological Gradient")
plt.axis("off")
```

```python
tophat = cv2.morphologyEx(binary, cv2.MORPH_TOPHAT, kernel)
blackhat = cv2.morphologyEx(binary, cv2.MORPH_BLACKHAT, kernel)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(tophat, cmap="gray")
plt.title("Top Hat")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(blackhat, cmap="gray")
plt.title("Black Hat")
plt.axis("off")
```
